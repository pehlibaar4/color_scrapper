```python
import os

import undetected_chromedriver as uc
from bs4 import BeautifulSoup
from requestium import Keys, Session
from selenium.webdriver.common.by import By
import random
import time
import csv
```

```python
proxy = { 'http': 'http://intgw-hyd.bhel.in:8080', 'https': 'http://intgw-hyd.bhel.in:8080' }
def switch_to_requests(s):
    s.transfer_driver_cookies_to_session()
    if os.environ == "Windows":
        s.proxies.update(proxy)
```

```python
base_url = r"https://www.canva.com/colors/color-palettes/"
```

```python
browser = uc.Chrome(version_main=105)
s = Session(driver=browser)
```

```python
page = None
for page_no in range(2, 100):
    time.sleep(random.uniform(3, 6))
    next_url = f"{base_url}{page if page else ''}"
    page = f"page/{page_no}/"

    print(next_url)

    s.driver.get(next_url)
    source = s.driver.page_source
    soup = BeautifulSoup(source)

    palette_cards = soup.find_all('span', class_="B1lyeQTjtngYnsfgvOZaoQ== reOL7E3gT30b7lAry+yThQ==")

    for palettes in palette_cards:
        colors_list = palettes.find_all('span', class_="Rbq9ca+ur-yqXjjFlvuVxg==")
        colors =  [color['style'].split(':')[1] for color in colors_list]
        print(', '.join(colors))

        with open("Canva_pallettes.csv", 'a+', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(colors)
```
